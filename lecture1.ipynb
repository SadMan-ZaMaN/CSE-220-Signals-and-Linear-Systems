```python
#Loop

for i in range(2,5):
    print(i)  # prints 2, 3, 4

for i in range(2,10,2):
    print(i, end = ' ')  # prints 2, 4, 6, 8


names = ['Sadman', 'Zaman', 'John', 'Doe']
for name in names:
    print(name)  # prints each name in the list

for index, name in enumerate(names):
    print(index, name)  # prints the index and name in the list
```

List

```python
numbers = [1, 2, 3, 4, 'two', True, 5.0]
print(numbers)  # prints the list


numbers.append(3)  # adds 3 to the end of the list

print(len(numbers))  

print(numbers[0])  # prints the first element of the list

print(numbers.count(3))  # prints the number of times 3 appears in the list

print(numbers.index(3))  # prints the index of the first occurrence of 3 in the list



```

```python
# Lists Mutable but Tuple are Immutable
numbers[0] = 10  # changes the first element of the list to 10
print(numbers)  # prints the modified list
```

# list comprehension

```python
# comprehension can make list in a single line

squared_numbers = [x**2 for x in range(10)]
print(squared_numbers)  # prints the list of squared numbers
```

list[start:stop:step]  
- start inclusive, stop exclusive, step default 1

```python
numbers = [1, 2, 3, 4, 'two', True, 5.0]
print(numbers[1:4])

print(numbers[-2::])  # prints the last two elements of the list

print(numbers[::-1])  # prints the list in reverse order
```

Dictionary  (key,value) pair

```python
ages = {'Sadman': 20, 'Zaman': 21, 'John': 22}  

print(ages)
print(ages.keys())  # prints the keys of the dictionary
print(ages.values())  # prints the values of the dictionary

print(ages.get('Sadman'))
print(ages.get('Messi'))        # None
```
