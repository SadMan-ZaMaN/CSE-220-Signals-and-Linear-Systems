# String

```python
# String


Message = "Hello, World"
print(Message)

print(len(Message))  # prints the length of the string

print(Message.upper())  # prints the string in uppercase

print(Message.split(','))  # splits the string into a list of words

name = "Sadman"
roll = 75
print('My name is', name, 'and my roll number is', roll)

print(f'{name} and {roll}')


# Concatinate
first_name = "Sadman"
last_name = "Zaman"
full_name = first_name + " " + last_name
print(full_name)

```

```python
# \\ is used to escape the backslash character in the string
path = 'F:\\CSE-220\\lecture1.ipynb'        
print(path)

# r is used to indicate that the string is a raw string and backslash will not be treated as an escape character
path = r'F:\CSE-220\lecture1.ipynb'        
print(path)
```
